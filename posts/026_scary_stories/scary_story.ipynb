Telling scary stories is one of the best parts of summer camp and sleepovers as a kid. Though I didn't love hearing about murderous psychopaths while sitting in the woods or in a dark basement, I have fond memories of cuddling with my friends and eating snacks while we all pretended not to be on the verge of tears. With the evolution of technology and social media, younger generations are slowly forgetting the art of telling scary stories. They may be tired of the classics with predictable endings, so I'd like to try mixing it up. Let's see if AI can help me write a story that strays away from typical tropes and is scary enough to make a grown man cry!!

**The LLM and I took turns contributing to the story, each writing one sentence at a time. Here's our final story!**

"The neighbor's security camera footage showed me arriving home at 6:47 PM, but I distinctly remember sitting in traffic until at least 7:30. I thought it was simply a misunderstanding, but when the figure turned to the camera, I saw my face. The timestamp burned in the corner—6:47 PM—and I watched myself unlock my front door with the same nervous habit of jiggling the key twice, something I'd never told anyone about. As the door opened, I saw my dog run towards the figure before sniffing and slowly backing away. The figure crouched down and extended my hand toward him in exactly the way I always did, but Rex's ears flattened against his skull, and a low whine escaped his throat that I'd never heard in eight years of owning him. The figure bent down and its body covered Rex from the camera. When it stood back up thirty seconds later, Rex was gone—no sound of claws on hardwood, no barking, just the figure wiping its hands on my jeans the way I do after petting him. Suddenly I heard my baby cry. The figure's head snapped toward the stairs with an unnatural speed, and then it smiled at the camera—my smile, but too wide, holding it for far too long as the crying upstairs grew more frantic. As I watched the camera, I began to cry, fearful to see what may have happened next. The footage continued—the figure climbing my stairs with footsteps that were somehow too light, too measured—and then my phone buzzed with a text from my own number: "Don't come home." I immediately dashed home, but once I'd gotten there, I wished I'd listened to the message. The front door was already open, swinging gently in the wind, and from inside I could hear my own voice singing the lullaby I always sang to Emma—but there were two voices, harmonizing. I noticed that my husband's shoes were at the door, but he was supposed to be at a conference. I stepped inside and the singing stopped immediately, replaced by the sound of something heavy being dragged across the floor upstairs, and then my husband's voice called down cheerfully, "Honey, you're home early—we were just playing a game." I looked in his arms, seeing a shape that mirrored the baby's. As my eyes adjusted to the dim hallway light, I realized the bundle in his arms was moving wrong—the joints bending in places they shouldn't, like it was being puppeteered from the inside. Frozen in place, I rattled my brain thinking of what might be happening, then it clicked. The conference wasn't until next week—I'd written it on the calendar myself—which meant my husband should have been at work, and whatever was holding that thing at the top of the stairs had been waiting here since 6:47 PM for the real me to come home. Behind me, I heard my own voice whisper, "We've been waiting for you," and when I spun around, I was staring at myself, still wearing the clothes I'd put on that morning, and in the reflection of her eyes I could see my husband standing at the top of the stairs—holding me."

Let's see why AI thinks the story was successful without having to follow typical tropes:

**Prompt:**

What do you think made this story work

**Response:**

- The uncanny valley of "self"
- Grounded, specific details
- Escalating stakes organically
- Ambiguity about what's happening
- The animals know
- That final reversal
- Collaborative building
