### General Thoughts

I've finished Joy Buolamwini's "Unmasking AI," and she did a great job of using the final sections to shift from diagnosis to action; here's where she really shows her vision for what needs to change.

I really appreciated her emphasis on accountability. She discusses the need for transparency in AI systems, arguing that people have a right to know when algorithms are making decisions about their lives. She also advocates for regulatory frameworks that would hold tech companies responsible for the harms their systems cause. The point isn't to stifle innovation, but ensure that vulnerable companies aren't paying the price for it.

I was inspired by her discussion of the Algorithmic Justice League. She founded this group to fight for equitable and accountable AI, and she frames this work as part of a larger movement for civil rights in the digital age. Just as previous generations fought for equal treatment under the law, we now need to fight for equal treatment by algorithms, especially seeing how they're increasingly governing our lives.

The book ends on a note that's both cautionary and hopeful. She paints an optimistic picture of what inclusive AI could look like. She recognizes the scale of the challenges ahead, but also believes change is possible if we collectively demand it. Reading this book has fundamentally shifted how I think about AI. It's not a neutral tool; it reflects our society's values and biases, and as a collective,  we have a responsibility to do better.

### Questions and Reflections:

- What questions remain?

I'm curious about what resistance from the tech industry would look like. How would companies respond to calls for greater accountability? I'd also like to know more about successful examples of algorithmic justice work!

- What will I take away from this book moving forward?
  
I'll approach AI systems with much more skepticism, especially when they're deployed in high-stakes contexts like criminal justice or hiring.

**This book is essential reading for anyone who wants to understand how AI is reshaping our world and who wants to ensure that transformation doesn't replicate the inequalities of the past.**
