We know AI is a useful tool for answering quick questions with facts from the internet and summarizing documents, but does it have the ability to think critically? If provided with bits of information, will it be able to combine those pieces to create the whole picture? I will attempt to play a classic game of Hangman with 2 LLMs and see which role they perform better in.

I tested both LLMs with the same 7-letter word, and though each had acquired enough letters to figure out the word, neither made a full guess, and instead guessed each letter 1-by-1, getting know more than 4 letters incorrect.

When I asked the LLM's to create the word, they both chose obscure words. I was able to guess most letters correctly, but noticed that as I was getting to the last blank spaces, none of the remaining letters would fit to make a coherent word.
