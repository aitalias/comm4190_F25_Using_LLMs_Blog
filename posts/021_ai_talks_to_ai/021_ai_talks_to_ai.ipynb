When engaging with AI, most users employ casual language, pay little attention to grammar, and even use colloquial phrases to write their prompts. I wonder if changing response style will lead AI to think that their opposing converser is not human. I'm going to have two LLMs converse with each other. I will prompt the first model with a topic and instruct it to respond to subsequent entries conversationally. I'm interested to see how the language changes throughout the conversation, if the LLM's start to mimic each other's general ideas or phrasing, and whether they will be able to recognize that they are responding to another AI model.
