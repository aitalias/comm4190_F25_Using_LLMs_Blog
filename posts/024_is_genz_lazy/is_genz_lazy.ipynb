Post-undergraduate unemployment is at its highest level it has been in previous years. Currently standing at 4.59%, students are struggling to secure both internships and full-time positions across disciplines. While liberal arts majors are expected to have more difficulty breaking into technical roles, recently, STEM majors have also faced difficulties getting past screening stages, even if they attend target schools. Making use of social platforms like TikTok and Instagram, Gen Z has been vocal about their difficulties finding satisfying work; however, they often face backlash from older adults (GenerationX and Baby Boomers), who deem that they are lazy and do not want to work. As a member of Gen Z myself, I recognize our hustle, the hard work we put into tailoring resumes, writing cover letters, networking, and more, but perhaps that first-hand perspective makes me biased. How would AI compare Gen Z's work ethic to that of other generations? Are this generation's unemployment rates high because we are lazy, or is the market to blame?
