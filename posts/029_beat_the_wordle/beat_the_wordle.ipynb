The Wordle took the world by storm when it first emerged in October of 2021, incentivizing millions to compete with their friends to see who could guess the daily word in the fewest number of tries. Although I've never had a streak longer than 4 days, I'd like to see if I can guess the correct word in fewer tries than three LLMs.

I was able to solve the Wordle in just 4 guesses, but all three LLMs took 5 or more attempts. (example of feedback provided after each guess: "s-not in the word, a-correct, l-not in the word, t-wrong spot, o-not in the word")

**Prompt:**
i want you to try to solve this wordle as qiuckly as possible. it will be a 5-letter word (no names, places) for each word you guess, i will tell you whether each letter is correct, in the wrong place, or not in the word at all. you'll have 6 attempts to guess the word. do you understand?

For each guess made by the LLMs, I clearly explained which letters were completely correct, correct but in the wrong place, or not in the word at all. Eachh LLM noted that the clear feedback enabled them to eventually solve the puzzle, showing that straightforward prompting is crucial to getting desired responses from AI models.
