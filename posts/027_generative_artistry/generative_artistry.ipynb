With generative AI, we're able to create images with almost any conceivable features with a simple request. It's interesting to see how different LLMs respond to the same prompt and how small tweaks to the prompt can change the result. Let's see how these LLMs respond to the same prompts! Which details are highlighted? Can the generated images reflect different artistic styles? Which LLM can generate an image that most closely resembles my vision?

**Prompt:** generate an image of a cow riding a bicycle

![](ss_027_02.png)

**Prompt:** make the bike pink, place a pink bow in the cows hair, place a pink lollipop in its hand, and make the bike on a tight rope

![](ss_027_03.png)

**Prompt:** can you adjust the animation style to look like claymation

![](ss_027_04.png)

**Prompt:** can the coloring be adjusted to appear more vibrant

![](ss_027_05.png)

**Prompt:** now make it fit this animation style
![](ss_27_06.png)

![](ss_027_07.png)

I was surprised to see that the LLM was able to incorporate all of my suggestions! In the past, LLMs have had trouble generating images that built off of previous prompts; when I asked for an addition to be made, it removed other details from prior generations. This was my first time experimenting with design style. It's interesting to see what animation/drawing style the LLM defaults to when generating images. Considering that this is a basic model with limited capabilities, I'm impressed that all of my requests could be fulfilled and that LLMs are able to mimic different illustrative styles!
